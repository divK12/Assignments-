```python
class Node:
    def __init__(self,tag='L',value=None,left=None,right=None):
        self.tag=tag
        self.value=value
        self.right=right
        self.left=left
        
    def find(self,v):
        if self.tag=='L':
            return v==self.value
        elif v<self.value and self.left:
            return(self.left.find(v))
        elif v>=self.value and self.right:
            return(self.right.find(v))
        else:
            return False
    
    def insert(self,v):
        if self.tag=='L':
            if self.value is None:
                self.value=v
            elif v<self.value:
                self.left=Node('L',v)
                self.right=Node('L',self.value)
                self.tag='I'
            elif v>self.value:
                self.right=Node('L',v)
                self.left=Node('L',self.value)
                self.value=v
                self.tag='I'
        elif self.value and v<self.value:
            if self.left is None:
                self.left=Node('L',v)
            else:
                self.left.insert(v)
        elif self.value and v>self.value:
            if self.right is None:
                self.right=Node('L',v)
            else:
                self.right.insert(v)
        elif not self.value:
            self.value=v
        
        return self
                        
            
        
    
    def minval(self):
        if not self.left :
            return(self.value)
        else:
            return(self.left.minval())
    
    def maxval(self):
        if not self.right:
            return self.value
        else:
            return self.right.maxval()

        
    def delete(self,v):
        if self.tag == 'I':
            if v == self.value:
                self.value = self.right.value
                self.left = self.left
                self.right = self.right.right
                self.rebalance()
            elif self.value and v < self.value:
                self.left.delete(v)
                self.rebalance()
            elif self.value and v >= self.value:
                self.right.delete(v)
                self.rebalance()
        elif self.tag == 'L':
             self.value = None  
        self.rebalance()
       

            

    def rebalance(self):
        if self.tag == 'I':
            if self.left is not None:
                self.left.rebalance()
            if self.right is not None:
                self.right.rebalance()
            if self.left is None and self.right is not None:
                self.value = self.right.minval()
            elif self.right is None and self.left is not None:
                self.value = self.left.maxval()
                self.left.value=None
     
        
    def inorder(self):
        result=[]
        if self.left:
            result+=self.left.inorder()
        if self.value is not None:
            result.append((self.tag,self.value))
        if self.right:
            result+=self.right.inorder()
        return result
            
    def __str__(self):
        return(str(self.inorder()))

        
        
```
