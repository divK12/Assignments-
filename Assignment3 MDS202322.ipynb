```python
#booking_id passed as parameters are taking int values 
import random
class BusBooking:
    def __init__(self):
        self.booking={}
        self.waiting={}
        
        self.seats={'W':list(range(1,21)),'A':list(range(1,21))}
        self.book_id=1
        
    def book(self,name,preference=None):
        sn=0
        if preference and preference.upper() in ['W','WINDOW']:
            pref='W'
        elif preference and preference.upper() in ['A','AISLE']:
            pref='A'
        else:
            pref=None
            
        if pref and self.seats[pref]:
            sn=self.seats[pref].pop(0)
            
        elif pref=='W' and self.seats['A']:
            sn=self.seats['A'].pop(0)
            pref='A'
            
        elif pref=='A' and self.seats['W']:
            sn=self.seats['W'].pop(0)
            pref='W'
            
        else:
            all_seats = self.seats['W'] + self.seats['A']
            if all_seats :
                pref = random.choice(['W', 'A'])
                sn = all_seats.pop(0)
                self.seats[pref].remove(sn)
            else:
                
                self.waiting[self.book_id]={'name': name, 'preference': pref}
                self.book_id+=1
                return self.book_id-1, f"WL-{len(self.waiting)}"
        
        self.booking[self.book_id]={'name':name,'preference':pref,'seat':sn}
        self.book_id+=1
        return self.book_id-1,f"{pref}{sn}"
    
    def cancel(self,booking_id):
        
        if booking_id in self.booking:
            sn=self.booking[booking_id]['seat']
            pref=self.booking[booking_id]['preference']
            del self.booking[booking_id]
            
            if self.waiting:
                wait_id,wdetails=self.waiting.popitem()
                wname=wdetails['name']
                self.booking[booking_id]={'name':wname,'preference':pref,'seat':sn}
                
                return True
            else:
                self.seats[pref].append(sn)
                return True
            
            
        elif booking_id in self.waiting:
            del self.waiting[booking_id]
            return True
        
        else:
            return False
    
   
    def status(self,booking_id):
        if booking_id in self.booking:
            pref=self.booking[booking_id]['preference']
            seat_number=self.booking[booking_id]['seat']
            name=self.booking[booking_id]['name']
            return (name,f"{pref}{seat_number}")
        
        elif booking_id in self.waiting:
            wname=self.waiting[booking_id]['name']
            return (wname,f"WL-{booking_id}")
            
        else:
            return None
   

    def __str__(self):
        booking_list = [(k, v['name'],v['preference'], v['seat']) for k, v in sorted(self.booking.items())]
        return str(booking_list)

                
```
